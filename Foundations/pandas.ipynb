# Series

```python
data = [100,102,104,432,312,532]

series = pd.Series(data, index = ['a','b','c','d','e','f'])

print(series)
#location
print(series.loc['a'])
print(series.loc['b'])
print(series.loc['c'])
#changing value in that location
series.loc['c'] = 200
print(series)
#locate by intiger position
print(series.iloc[2])
```

```python
#printing based on value

print(series[series>200])
```

```python
#Now we use dictionary

dict = {"icream":"chocolate","no of cones":1,"liked?":"yes"}
print(dict)
```

```python
series = pd.Series(dict)
print(series)
print(series.loc["icream"])
print(series.iloc[2])

```

# Dataframes


```python
data = {"Name":["Spongebob","Patrick","Squidward"],
        "Age":[20,30,50]
}

df = pd.DataFrame(data)
print(df)
print(df.loc[1])

df["Job"] = ["Cook","Waiter","Cheif"]
print(df)
```

```python
new_row = pd.DataFrame([{'Name':'Sandy',"Age":40,"Job":"waitres"}],
                       index = [3])
df = pd.concat([df, new_row])

print(df)
```
