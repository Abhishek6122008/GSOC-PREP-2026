# For Loops

```python
for i in range(1, 11):
    print(i)
```

```python
for i in reversed(range(1,11)):
    print(i)
```

## Jumps in For loops


```python
for i in range(1, 11, 3):
    print(i)
```

# Continue


```python
for i in range(1,21):
    if i == 13:
        continue
    else:
        print(i)
```

# Break

```python
for i in range(1,21):
    if i == 13:
        break
    else:
        print(i)
```
